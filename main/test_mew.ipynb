```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
#  100  1000  0-1 
# 
num_dimensions = 133  # 100 
time_length = 1000    # 1000 
frequency = 25        # 25Hz 25 

# 
# np.random.seed(42)  # 
data = np.zeros((num_dimensions, time_length), dtype=int)  #  0 

#  25  1
for i in range(num_dimensions):
    discharge_positions = np.random.choice(time_length, size=frequency, replace=False)  #  25 
    data[i, discharge_positions] = 1  #  1

#  1 
event_positions = [np.where(row == 1)[0] for row in data]

#  eventplot
plt.figure(figsize=(10, 6))
plt.eventplot(event_positions, colors='black', lineoffsets=1, linelengths=0.8)

 # 
# plt.title("Event Plot of 100-Dimensional 0-1 Sequence (25Hz, 1000ms)")
plt.xlabel("Time (ms)")
plt.ylabel("Neuron index")

#  0-1000 
plt.xlim(0, time_length)

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
x = np.array([1, 3, 5, 7, 9, 11, 13])
y = np.array([0.11, 0.33, 0.55, 0.71, 0.9, 1.05, 1.2])
plt.rcParams.update({'font.size': 15})
# linear regression and plot the line 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
x_fit = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
y_fit = model.predict(x_fit)
plt.plot(x_fit, y_fit, color='black', linestyle='-', linewidth=2, label='Linear Fit')

# 
# 
plt.scatter(x, y, facecolors='none', edgecolors='black', s=100)

# 
plt.xlabel(r"$-V_E V_I (mV^2)$")
plt.ylabel(r"$-V_{sc} (mV)$")

# 
plt.show()
```
