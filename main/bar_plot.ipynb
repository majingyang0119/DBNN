```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def gabor_filter(frequency, theta, gamma, sigma_x, sigma_y, size):
    x, y = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))
    
    x_theta = x * np.cos(theta) + y * np.sin(theta)
    y_theta = -x * np.sin(theta) + y * np.cos(theta)
    
    g = np.exp(-((x_theta**2) / (2 * sigma_x**2) + (gamma**2 * y_theta**2) / (2 * sigma_y**2))) * np.cos(2 * np.pi * frequency * x_theta)
    return g

frequency = 20
theta = np.pi / 6 * 3
gamma = 2
sigma_x = 0.2
sigma_y = 0.2
size = 31

gabor = gabor_filter(frequency, theta, gamma, sigma_x, sigma_y, size)

plt.imshow(gabor, cmap='gray')
# plt.title('Gabor Filter')
plt.axis('off')
# plt.show()
plt.savefig('figure/gabor_90_0.2.png', bbox_inches='tight', dpi=800)
```
