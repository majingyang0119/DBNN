```python
import numpy as np
import matplotlib.pyplot as plt
# Dimensions and parameters
N = 100
T = 1000
firing_rate = 20

# Initialize dataset
data = np.zeros((N, T), dtype=int)

# Set random seed for reproducibility
np.random.seed(42)

# Randomly pick 20 dimensions to have identical firing times
selected_indices = np.random.choice(N, 20, replace=False)
common_firing_indices = np.random.choice(T, firing_rate, replace=False)

# Set firing times for the selected dimensions
for i in selected_indices:
    data[i, common_firing_indices] = 1

# Set firing times for the other dimensions
for i in range(N):
    if i not in selected_indices:
        unique_firing_indices = np.random.choice(T, firing_rate, replace=False)
        data[i, unique_firing_indices] = 1
```

```python
# Create the raster plot with color coding
plt.figure(figsize=(12, 8))
for i in range(N):
    spike_times = np.where(data[i] == 1)[0]  # Find indices where there is a spike
    if i in selected_indices:
        plt.eventplot(spike_times, lineoffsets=i, linelengths=1.8, color='black')  # Plot in red for selected
        
    else:
        plt.eventplot(spike_times, lineoffsets=i, linelengths=1.8, color='black')  # Plot in black for others

# plt.title('pattern 1')
# plt.xlabel('Time')
# plt.ylabel('Dimension')
# plt.xlim([0, T])
# plt.ylim([0, N])
plt.axis('off')
# plt.show()
plt.savefig('figure/100D_raster_plot1_new.png', dpi=800, bbox_inches='tight')
```
