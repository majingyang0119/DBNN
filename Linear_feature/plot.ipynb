```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 30})
# 
matrix_size = 10
# matrix = np.random.uniform(-1.1, 1.1, size=(matrix_size, matrix_size))
matrix = np.random.randn(matrix_size, matrix_size) * 0.15

symmetric_matrix = (matrix + matrix.T) / 2  # 
symmetric_matrix[2, 0] = 0.85
symmetric_matrix[3, 1] = 0.60
symmetric_matrix[7, 6] = 0.66
symmetric_matrix[9, 8] = 0.72
symmetric_matrix[6, 2] = -1.16
symmetric_matrix[9, 0] = -1.25

# 
mask = np.triu(np.ones_like(symmetric_matrix, dtype=bool))

#  NaN
symmetric_matrix[mask] = np.nan

# 
plt.figure(figsize=(8, 6))
plt.imshow(symmetric_matrix, cmap='coolwarm', interpolation='nearest', vmin=-1.3, vmax=1.3)
plt.colorbar(label=r'$\kappa$ value')
plt.xticks(np.arange(matrix_size))
plt.xlabel('input index')
plt.ylabel('input index')
# plt.title('Bilinear matrix distribution of DLIF model')
plt.yticks(np.arange(matrix_size))
plt.gca().set_xticklabels(np.arange(1, matrix_size + 1))
plt.gca().set_yticklabels(np.arange(1, matrix_size + 1))
plt.savefig('figure/kappa_matrix_corr.png', dpi=800, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
# 
matrix_size = 10
matrix = np.random.uniform(-1.1, 1.1, size=(matrix_size, matrix_size))
symmetric_matrix = (matrix + matrix.T) / 2  # 

# 
mask = np.triu(np.ones_like(symmetric_matrix, dtype=bool))

#  NaN
symmetric_matrix[mask] = np.nan

# 
plt.figure(figsize=(8, 6))
plt.imshow(symmetric_matrix, cmap='coolwarm', interpolation='nearest', vmin=-1, vmax=1)
plt.colorbar(label=r'$\kappa$')
plt.xticks(np.arange(matrix_size))
plt.yticks(np.arange(matrix_size))
plt.gca().set_xticklabels(np.arange(1, matrix_size + 1))
plt.gca().set_yticklabels(np.arange(1, matrix_size + 1))
plt.savefig('figure/kappa_matrix_bcg.png', dpi=800, bbox_inches='tight')
plt.show()
```
