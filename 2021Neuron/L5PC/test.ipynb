```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Define the network
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3, 3)
        self.fc2 = nn.Linear(3qq, 3)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Create the network
net = Net()

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.01)

# Define the inputs and labels
inputs = torch.Tensor([[1, 1, 0], [0, 0, 1], [1, 0, 1], [0, 1, 0], [0, 1, 1], [1, 0, 0]])
labels = torch.LongTensor([0, 0, 1, 1, 2, 2])
labels_one_hot = F.one_hot(labels, num_classes=3)

# Train the network
for epoch in range(1000):
    optimizer.zero_grad()
    outputs = net(inputs)
    loss = criterion(outputs, labels_one_hot.float())
    loss.backward()
    optimizer.step()
    # Print loss every 100 epochs
    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')

# Test the network
outputs = net(inputs)
_, predicted = torch.max(outputs, 1)
print('Predicted: ', predicted.numpy())
print('Labels: ', labels.numpy())
```
