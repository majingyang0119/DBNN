```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 15})
```

```python
NAR = [2, 1, 0.7, 0.3, 0]
VE = np.array([[0.67, 0.58, 0.52], [0.705, 0.73, 0.811], [0.9, 0.84, 0.82], [0.96, 0.92, 0.912], [0.99, 0.97, 0.972]])
VE_mean = np.mean(VE, axis=1)
yerr = np.zeros((2, 5))
for i in range(yerr.shape[1]):
    yerr[0][i] = np.max(VE[i]) - np.mean(VE[i])
    yerr[1][i] = np.mean(VE[i]) - np.min(VE[i])

VE_new = np.array([[0.77, 0.7, 0.66], [0.86, 0.91, 0.94], [0.92, 0.96, 0.94], [0.98, 0.96, 0.95], [0.99, 0.97, 0.972]])
VE_mean_new = np.mean(VE_new, axis=1)
yerr_new = np.zeros((2, 5))
for i in range(yerr_new.shape[1]):
    yerr_new[0][i] = np.max(VE_new[i]) - np.mean(VE_new[i])
    yerr_new[1][i] = np.mean(VE_new[i]) - np.min(VE_new[i])
```

```python
plt.figure(figsize=(6,4))
# plt.plot(NAR, VE_mean, marker = 'o', markersize=4, color = 'red', linewidth = 0.1, linestyle = '-')
plt.plot(NAR, VE_mean_new, marker = 'o', markersize=4, color = 'red', linewidth = 0.1, linestyle = '-')
# plt.errorbar(NAR, VE_mean, yerr = yerr, c='blue', label='old')
plt.errorbar(NAR, VE_mean_new, yerr = yerr_new, c='green', label='new')
plt.xlabel('NAR')
plt.ylabel('variance explained')
# plt.legend()
plt.savefig('figure/VE-NAR_new_single.png', bbox_inches='tight')
```
